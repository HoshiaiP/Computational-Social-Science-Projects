#### d. Check for duplicate columns

There are a lot of columns in this data frame. Let's see if there are any are duplicates. Note that Pandas will not allow them to have the same exact column name, so they will likely be distinct on column name but will be copies otherwise.

```python
#
# Check for duplicate columns
#-----------
# I used Google to figure this out, and adapted this example for our purposes:  
# source: https://thispointer.com/how-to-find-drop-duplicate-columns-in-a-dataframe-python-pandas/ 
def getDuplicateColumns(df):
    '''
    Get a list of duplicate columns.
    It will iterate over all the columns in dataframe and find the columns whose contents are duplicate.
    :param df: Dataframe object
    :return: List of columns whose contents are duplicates.
    '''
    duplicateColumnNames = set()
    # Iterate over all the columns in dataframe
    for x in range(df.shape[1]):
        # Select column at xth index.
        col = df.iloc[:, x]
        # Iterate over all the columns in DataFrame from (x+1)th index till end
        for y in range(x + 1, df.shape[1]):
            # Select column at yth index.
            otherCol = df.iloc[:, y]
            # Check if two columns at x 7 y index are equal
            if col.equals(otherCol):
                duplicateColumnNames.add(df.columns.values[y])
    return list(duplicateColumnNames)

duplicateColumnNames = list(getDuplicateColumns(diabetes))
print('Duplicate Columns are as follows: ')
duplicateColumnNames
```

```python
#
# drop columns from duplicates list
#-----------
# now drop list of duplicate features from our df using the .drop() method
diabetes = diabetes.drop(columns=duplicateColumnNames) 
```

Now, we will partition our data to prepare it for the training process. Ultimately we want to use a 60% train—20% validation—20% test in this case. More data in the training set lowers bias, but then increases variance in the validation/test sets. Balancing between bias and variance with choice of these set sizes is important as we want to ensure that there is enough data to train on to get good predictions, but also want to make sure our hold-out sets are representative enough.

Standardization takes place after the training/validation/test split to avoid data leakage from the test data into training data.

```python
#
# Standardization
#-----------
# Given that we want to only standardize non-dichotomous variables, we need to find a 
# solution that will loop over only the columns we want to standardize. The code below
# identifies all non-dichotomous variables in our dataset and only standardizes those.

# load library and create instance of Standard Scaler 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()


# identify non-dichotomous columns we want to transform
columns = list(X_test.select_dtypes(include=['number']).loc[:, X_test.nunique() > 2])

# use loop to transform training data for only columns we want to transform
for column in columns:
    X_train[column] = scaler.fit_transform(X_train[column].values.reshape(-1, 1)).flatten()

# use loop to transform validation data for only columns we want to transform
for column in columns:
    X_validate[column] = scaler.fit_transform(X_validate[column].values.reshape(-1, 1)).flatten()

# use loop to transform test data for only columns we want to transform
for column in columns:
    X_test[column] = scaler.fit_transform(X_test[column].values.reshape(-1, 1)).flatten()
```

### 4.1 Describe models

Detail the basic logic and assumptions underlying each model, its pros/cons, and why it is a plausible choice for this problem.

**MODEL DESCRIPTION(S):** 
The three models I use are OLS, Ridge, and LASSO.

Model 1: OLS
Basic logic: finding the best-fitting line or hyperplane such that the sum of the squared residuals is the smallest.

Assumptions: 
    
    Independence: data points are independent from each other.
    
    Linearity: the relationship between predictors and outcomes is linear.
    
    Homoscedasticity: the variance of residuals do not change with the predictors.
    
    No Multicollinearity: the predictors are not perfectly correlated with each other. 
    
    Normality: errors are normally distributed.

Pros: easy interpretation and low computational requirement

Cons: assumptions of independence and no multicollinearity are often violated in social scientific applications; prone to overfitting when there are large number of predictors relatve to observations. 

Model 2: Ridge

Basic logic: adding a shrinkage penalty to shrink overly complex models, in order to prevent overfitting. When the penalty is small, the coefficient will approach zero. 

Assumptions: same as OLS, except that with the tuning parameter the assumption of no multicollinearity can be loosened. 

Pros: performs better than OLS when predictors are correlated 

Cons: keeps all predictors in the model, which becomes very dense and difficult to interpret.

Model 3: Least Absolute Shrinkage and Selection Operator (LASSO)

Basic logic: similar to Ridge, except that the shrinkage penalty is the absolute value of coefficients. 

Assumptions: same as Ridge

Pros: the shrinkage penalty allows some coefficients to be set as exactly zero, effectively dropping some predictors. This makes models more interpretable compared to Ridge.

Cons: might arbitrarily select one out of a set of correlated predictors.

For this problem, it is unclear which predictors of diabetes rates are more relevant, and what the level of multicollinearity is , so it makes sense to test out OLS, Ridge, and LASSO and compare their results.

1. What is bias-variance tradeoff? Why is it relevant to machine learning problems like this one?

The bias-variance tradeoff is that as a model complexity increases, bias (how much estimates deviate from actual value) decreases and variance (how sensitive a model is to changes in the training data) increases. Conversely, as model complexity decreases, bias increases and variance decreases. The bias-variance trade-off is relevant to this problem because in social science we care about both model accuracy and generalizability, therefore, we want both low bias and low variance. Since that is not possible, we need to find an appropriate balance between the two.  

2. Define overfitting, and why it matters for machine learning. How can we address it?

Overfitting is when a model captures too much of the underlying patterns of the data, including idiosyncrasies of the training data, which causes high performance on the training data but poor generalization to new data. Some form of variable selection, including shrinkage methods such as Ridge and LASSO regressions, are ways to address overfitting by reducing model complexity.
